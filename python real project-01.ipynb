```python
#Finding the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.
```

```python
channel_counts = data.groupby(['channel_title'])['video_id'].count()

# sort the channels by video count in decreasing order
top_channels = channel_counts.sort_values(ascending=False).head(10)
bottom_channels = channel_counts.sort_values(ascending=True).head(10)

# display the resulting top and bottom channels
print('Top 10 channels with highest video count:\n', top_channels)
print('\nBottom 10 channels with lowest video count:\n', bottom_channels)
```

```python
## Finding the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.
```

```python

# find the title of the video with maximum likes
max_likes_index = data['likes'].idxmax()
max_likes_title = data.loc[max_likes_index, 'title']

# find the title of the video with minimum likes
min_likes_index = data['likes'].idxmin()
min_likes_title = data.loc[min_likes_index, 'title']

print("Video with maximum likes:", max_likes_title)
print("Video with minimum likes:", min_likes_title)
```

```python
## inferences
##To find the title of the video with the maximum number of likes, used the idxmax() function and idxmin()
## to get the index of the row with the maximum value in the "likes" column, and then used the loc function to retrieve the title of the video from the "title" column at that index.
## I got BTS () 'Dynamite' Official MV as highest likes.
## I got the Kim Kardashian's Must-See Moments on "Saturday Night Live" | E! News
```

```python
##Find the title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes 
```

```python
# find the title of the video with maximum dislikes
max_dislikes_index = data['dislikes'].idxmax()
max_dislikes_title = data.loc[max_dislikes_index, 'title']

# find the title of the video with minimum dislikes
min_dislikes_index = data['dislikes'].idxmin()
min_dislikes_title = data.loc[min_dislikes_index, 'title']

print("Video with maximum dislikes:", max_dislikes_title)
print("Video with minimum dislikes:", min_dislikes_title)
```
